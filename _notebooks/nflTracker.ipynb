```python
import sqlite3

# Define a function to create the database table if it doesn't exist
def create_table_if_not_exists():
    conn = sqlite3.connect('nfl_tracker.db')
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS nfl_games
                 (id INTEGER PRIMARY KEY AUTOINCREMENT,
                  home_team TEXT,
                  away_team TEXT,
                  home_team_win_percentage REAL,
                  away_team_win_percentage REAL)''')
    conn.commit()
    conn.close()

# Define the function to add NFL games to the tracker database
def add_game_to_database(home_team, away_team, home_win_percentage, away_win_percentage):
    conn = sqlite3.connect('nfl_tracker.db')
    c = conn.cursor()
    c.execute("INSERT INTO nfl_games (home_team, away_team, home_team_win_percentage, away_team_win_percentage) VALUES (?, ?, ?, ?)",
              (home_team, away_team, home_win_percentage, away_win_percentage))
    conn.commit()
    conn.close()

# Define the function to retrieve all NFL games from the tracker database
def get_all_games_from_database():
    conn = sqlite3.connect('nfl_tracker.db')
    c = conn.cursor()
    c.execute("SELECT home_team, away_team, home_team_win_percentage, away_team_win_percentage FROM nfl_games")
    games = c.fetchall()
    conn.close()
    return games

# Define the function to add NFL games to the tracker
def add_game():
    while True:
        home_team = input("Enter the home team: ")
        away_team = input("Enter the away team: ")
        home_win_percentage = float(input(f"Enter the win percentage for {home_team}: "))
        away_win_percentage = float(input(f"Enter the win percentage for {away_team}: "))

        add_game_to_database(home_team, away_team, home_win_percentage, away_win_percentage)
        print("Game added successfully!")

        another_game = input("Do you want to add another game? (yes/no): ")
        if another_game.lower() != 'yes':
            break

# Main function to run the program
def main():
    # Create the database table if it doesn't exist
    create_table_if_not_exists()

    # Call the add_game function to add games
    add_game()

    # Output the list of tracked NFL games
    print("\nNFL Games Being Tracked:")
    nfl_games = get_all_games_from_database()
    for i, game in enumerate(nfl_games, 1):
        print(f"{i}. Home Team: {game[0]}, Away Team: {game[1]}, Home Win Percentage: {game[2]}, Away Win Percentage: {game[3]}")

# Call the main function to start the program
if __name__ == "__main__":
    main()

```
